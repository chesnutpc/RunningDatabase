## Motivation for the Database Project

This project aims to help runners track and analyze their running activities, providing valuable insights into their performance, health, and progress. The primary motivation is to create an intuitive and efficient database for managing personal running data. This database will enable users to easily retrieve and visualize information about their runs, such as time, weather conditions, shoes worn, run length, pace, and heart rate. By centralizing this data, users can make informed decisions to improve their running experience, optimize training, and remain motivated to achieve their running goals.


## Involvement of Multiple Relations and Data Sources

The Running Log Database will involve multiple related data tables to store the various aspects of each run. Relations will exist between tables for storing data on individual runs, weather conditions, shoe information, and run waypoint records that contain detailed metrics such as GPS coordinates, heart rate, pace, GPS coordinates and distance. ## Platform Choices and Justification

1. **Database Creation:** MySQL will be used for the creation and management of the database. MySQL is an ideal choice for this project due to its simplicity and ease of integration with Python and Jupyter Notebooks. It is well-suited for personal applications and is capable of handling the expected data volume efficiently.  A Jupyter notebook will be authored to create all tables with integrity constraints.  Python scripts will be written to parse the csv files and import them into the database.

2. **Data Querying and Visualization:** For querying and presenting the data, Plotly Dash will be employed. Plotly Dash is a powerful Python framework for building interactive web applications or embedding in Jupyter Notebooks, which will enable users to visualize their running data with interactive charts and graphs. The choice of Dash allows for the creation of a user-friendly graphical user interface to query and visualize data, enhancing the project's usability.


## Learning Outcomes

The key learning outcomes for this project are as follows:

2. **Data Import and Transformation:** Learning how to import data from external CSV files generated by my iPhone and transforming it into a format suitable for storage in the database.

3. **Data Querying and Visualization:** Gaining expertise in querying the database to retrieve specific information and visualizing the data using Plotly Dash. This includes creating interactive and user-friendly visualizations.
